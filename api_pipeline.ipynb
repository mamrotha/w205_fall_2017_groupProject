#Crypto Coin Market Cap Pipeline

```python
import requests
import pandas as pd
import numpy as np
import datetime
```

```python
r = requests.get("https://api.coinmarketcap.com/v1/ticker/?limit=0")
ticker = pd.DataFrame(r.json())
ticker['last_updated'] = pd.to_datetime(ticker['last_updated'],unit='s')
ticker
ticker.to_csv(("ticker_{}".format(max(ticker["last_updated"]))).replace(" ", "_"))
```
